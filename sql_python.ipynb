````python
# Criando uma tabela Movies na base de dados
try:
    create_table_query = '''
        CREATE TABLE Movies (
        MovieID SERIAL NOT NULL,
        Title VARCHAR(255) NOT NULL,
        Genre VARCHAR(100),
        ReleaseYear INTEGER CHECK (ReleaseYear BETWEEN 1888 AND 2100),
        Rating DECIMAL(2, 1),
        Description TEXT,
        CreatedAt TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        PRIMARY KEY (MovieID)
            );
        '''
        # Executa o comando de criação
    cursor.execute(create_table_query)
    connection.commit()
except:
    print('Tabela ja existe!')
````

````python
# Criando uma tabela Users na base de dados
try:
    create_table_query = '''
        CREATE TABLE Users (
        UserID Serial NOT NULL,
        Name VARCHAR(100),
        PhoneNumber VARCHAR(20) UNIQUE,
        CreatedAt TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        PRIMARY KEY (UserID)
    );
        '''
        # Executa o comando de criação
    cursor.execute(create_table_query)
    connection.commit()
except:
    print('Tabela ja existe!')
    
````

````python
# Criando uma tabela Recommendations na base de dados
try:
    create_table_query = '''
        CREATE TABLE Recommendations (
        RecommendationID Serial NOT NULL,
        UserID INT,
        MovieID INT,
        RecommendedAt TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        PRIMARY KEY (RecommendationID),
        FOREIGN KEY (UserID) REFERENCES Users(UserID),
        FOREIGN KEY (MovieID) REFERENCES Movies(MovieID)
    );
        '''
        # Executa o comando de criação
    cursor.execute(create_table_query)
    connection.commit()
except:
    print('Tabela ja existe!')
````

````python
# Criando uma tabela Interactions na base de dados
try:
    create_table_query = '''
        CREATE TABLE Interactions (
        InteractionID Serial NOT NULL,
        UserID INT,
        Message TEXT,
        InteractionTime TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        PRIMARY KEY (InteractionID),
        FOREIGN KEY (UserID) REFERENCES Users(UserID)
    );
        '''
        # Executa o comando de criação
    cursor.execute(create_table_query)
    connection.commit()
except:
    print('Tabela ja existe!')
````

````python
# Criar o motor de conexão
engine = create_engine(connection_string, echo=False)
````
